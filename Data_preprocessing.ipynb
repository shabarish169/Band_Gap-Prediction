```python
import rdkit.Chem.Descriptors

# Get all attributes of the module
all_attributes = dir(rdkit.Chem.Descriptors)

# Filter out only functions
functions = [attr for attr in all_attributes if callable(getattr(rdkit.Chem.Descriptors, attr))]

# Print the list of functions
print("Functions in the module:", functions)
```

```python
descriptor_results = []

# Apply each function to the 'mol' column and store the results in the list
for f in functions:
  try:
    descriptor_func = getattr(rdkit.Chem.Descriptors, f)  # Dynamically access function from Descriptors module
    descriptor_results.append(df['mol'].apply(lambda x: descriptor_func(x)))
  except:
    continue

# Concatenate all the descriptor results into a single DataFrame
descriptor_df = pd.concat(descriptor_results, axis=1)

# Concatenate the original DataFrame with the descriptor DataFrame
df = pd.concat([df, descriptor_df], axis=1)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# train_df = train_df.sample(frac=1, axis=1, random_state=42)     # This is to randomly shuffle columns
# X = df.select_dtypes(include='number')
X = df
# Perform PCA
pca = PCA(n_components=X.shape[1] - 1)  # Choose the number of components
X_pca = pca.fit_transform(X)
```

```python
len(X_pca[0])
```

```python
cumulative_explained_variance_threshold = 0.9995

cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)
num_components_exceeding_threshold = np.argmax(cumulative_explained_variance >= cumulative_explained_variance_threshold) + 1

# Capture the chosen columns based on the threshold
chosen_columns = df.columns[:num_components_exceeding_threshold].tolist()

# Create a DataFrame with PCA components
pca_components_df = pd.DataFrame(data=X_pca[:, :num_components_exceeding_threshold], columns=[f'pca_{i+1}' for i in range(num_components_exceeding_threshold)])
```

```python
print("\nPCA Components DataFrame:")
pca_components_df
```
